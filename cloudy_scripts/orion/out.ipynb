```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('pnpn.con', sep='\t')
wl = data['#Cont  nu']
flux = data['total']
data
```

```python
plt.plot(wl, flux)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Wavelength (in microns)')
plt.ylabel('$\\nu L_\\nu$ (erg/s)')
plt.xlim(1e-3, 1e7)
```
