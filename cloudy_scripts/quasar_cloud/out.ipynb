```python
import matplotlib.pyplot as plt
import pandas as pd
from numpy import log
```

```python
data = pd.read_csv('blrblr.con', sep='\t')
wl = data['#Cont  nu']
flux = data['total']
data
```

```python
plt.plot(wl, flux)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Wavelength (in microns)')
plt.ylabel('$L_\\nu$ (erg/cm$^2$/s)')
plt.xlim(1e-1, 1e1)
plt.ylim(1.5e7, 1e11)
```
